```python
# Plot a boxplot to compare the covid child cases, adolescent case, adult case and elderly case.
# The notch = True attribute creates the notch format to the box plot
# patch_artist = True fills the boxplot with colors, different colors for different boxes.
# The vert = 0 attribute creates horizontal box plot. 
# Import libraries
import matplotlib.pyplot as plt
import numpy as np
 
# Creating dataset
np.random.seed(10)
data_child = np.random.normal(32.545048, 79.708481, 600)
data_adolescent = np.random.normal(16.024501 ,41.433625, 600)
data_adult = np.random.normal(165.724037, 419.303178, 600)
data_elderly = np.random.normal(21.800636, 48.383413, 600)
data = [data_child, data_adolescent, data_adult, data_elderly]
 
fig = plt.figure(figsize =(8, 6))
ax = fig.add_subplot(111)
 
# Creating axes instance
Population = ax.boxplot(data, patch_artist = True,
                notch ='True', vert = 0)
 
colors = ['#0000FF', '#00FF00',
          '#FFFF00', '#FF00FF']
 
for patch, color in zip(Population['boxes'], colors):
    patch.set_facecolor(color)
 
# changing color and linewidth of
# whiskers
for whisker in Population['whiskers']:
    whisker.set(color ='#8B008B',
                linewidth = 1.5,
                linestyle =":")
 
# changing color and linewidth of
# caps
for cap in Population['caps']:
    cap.set(color ='#8B008B',
            linewidth = 2)
 
# changing color and linewidth of
# medians
for median in Population['medians']:
    median.set(color ='red',
               linewidth = 3)
 
# changing style of fliers
for flier in Population['fliers']:
    flier.set(marker ='D',
              color ='#e7298a',
              alpha = 0.5)
     
# x-axis labels
ax.set_yticklabels(['data_child', 'data_adolescent',
                    'data_adult', 'data_elderly'])
 
# Adding title
plt.title("Covid cases between child, adolescent, adult and elderly")
 
# Removing top axes and right axes
# ticks
ax.get_xaxis().tick_bottom()
ax.get_yaxis().tick_left()
     
# show plot
plt.show(Population)
```

```python
# second method is to use mean to find the na values by column
import pandas as pd 
import numpy as np
 
# Importing the SimpleImputer class
from sklearn.impute import SimpleImputer
 
# Imputer object using the mean strategy and
# missing_values type for imputation
imputer = SimpleImputer(missing_values = np.nan,
                        strategy ='mean')
 
deaths_age = pd.read_csv('https://raw.githubusercontent.com/MoH-Malaysia/covid19-public/main/epidemic/deaths_age.csv')
 
deaths_age['perc_0_4'] = deaths_age['perc_0_4'].fillna(deaths_age['perc_0_4'].mean())
deaths_age['perc_5_11'] = deaths_age['perc_5_11'].fillna(deaths_age['perc_5_11'].mean())
deaths_age['perc_12_17'] = deaths_age['perc_12_17'].fillna(deaths_age['perc_12_17'].mean())
deaths_age['perc_18_29'] = deaths_age['perc_18_29'].fillna(deaths_age['perc_18_29'].mean())
deaths_age['perc_30_39'] = deaths_age['perc_30_39'].fillna(deaths_age['perc_30_39'].mean())
deaths_age['perc_40_49'] = deaths_age['perc_40_49'].fillna(deaths_age['perc_40_49'].mean())
deaths_age['perc_50_59'] = deaths_age['perc_50_59'].fillna(deaths_age['perc_50_59'].mean())
deaths_age['perc_60_69'] = deaths_age['perc_60_69'].fillna(deaths_age['perc_60_69'].mean())
deaths_age['perc_70_79'] = deaths_age['perc_70_79'].fillna(deaths_age['perc_70_79'].mean())
deaths_age['perc_80_+'] = deaths_age['perc_80_+'].fillna(deaths_age['perc_80_+'].mean())
deaths_age
```

```python
# Import libraries
import matplotlib.pyplot as plt
import numpy as np
 
# Creating dataset
np.random.seed(10)
data_rtk_ag = np.random.normal(32058.461646, 36036.336268, 45000)
data_pcr = np.random.normal(27695.207810, 22063.055485, 45000)

data_test = [data_rtk_ag, data_pcr]
 
fig = plt.figure(figsize =(10, 7))
ax = fig.add_subplot(111)
 
# Creating axes instance
tests_plot = ax.boxplot(data_test, patch_artist = True,
                notch ='True', vert = 0)
 
colors = ['#0000FF', '#00FF00',
          '#FFFF00', '#FF00FF']
 
for patch, color in zip(tests_plot['boxes'], colors):
    patch.set_facecolor(color)
 
# changing color and linewidth of
# whiskers
for whisker in tests_plot['whiskers']:
    whisker.set(color ='#8B008B',
                linewidth = 1.5,
                linestyle =":")
 
# changing color and linewidth of
# caps
for cap in tests_plot['caps']:
    cap.set(color ='#8B008B',
            linewidth = 2)
 
# changing color and linewidth of
# medians
for median in tests_plot['medians']:
    median.set(color ='red',
               linewidth = 3)
 
# changing style of fliers
for flier in tests_plot['fliers']:
    flier.set(marker ='D',
              color ='#e7298a',
              alpha = 0.5)
     
# x-axis labels
ax.set_yticklabels(['rtk-ag', 'pcr',])
 
# Adding title
plt.title("COVID-19 Test rtk-ag vs pcr box plot")
 
# Removing top axes and right axes
# ticks
ax.get_xaxis().tick_bottom()
ax.get_yaxis().tick_left()
     
# show plot
plt.show(tests_plot)
```

```python
# based on the boxplots above, we can conclude that rtk-ag antigen rapid test is more commonly used than pcr test for COVID-19 testing.
```
